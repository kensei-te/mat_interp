# 付属 example data生成
example1_train: ErCo2 train  
example2_train: Fe3Ga4 train  
example4_train: BiS2 HPA FS train  

example3 は、著作権があるかもしれないのでそのまま

```python
import pandas as pd
df_source1 = pd.read_csv("example1_train.csv")
df_source1
```

```python
df_temp1a = pd.read_csv("example1_train.csv", usecols=["Temperature (K)", "Field (Oe)", "Long Moment (emu)"])
df_temp1a
df_temp1b = df_temp1a.reindex(columns=["Temperature (K)","Field (Oe)","Long Moment (emu)"])
df_temp1b
df_temp1b.columns=["X1_demo1", "X2_demo1", "Y_demo1"]
df_temp1b.to_csv("traindata_demo1.csv", index=False)
df_temp1b
```

```python
df_source2 = pd.read_csv("example2_train.csv")
df_source2

df_temp2a = pd.read_csv("example2_train.csv", usecols=["Temperature (K)", "Field (Oe)", "Long Moment (emu)"])
df_temp2a
df_temp2b = df_temp2a.reindex(columns=["Temperature (K)","Field (Oe)","Long Moment (emu)"])
df_temp2b
df_temp2b.columns=["X1_demo2", "X2_demo2", "Y_demo2"]
df_temp2b.to_csv("traindata_demo2.csv", index=False)
df_temp2b
```

```python
df_source4 = pd.read_csv("example4_train.csv")
df_source4

df_temp4a = pd.read_csv("example4_train.csv", usecols=["prim_axis", "sec_axis", "value"])
df_temp4a
df_temp4a.columns=["X1_demo4", "X2_demo4", "Y_demo4"]
df_temp4a.to_csv("traindata_demo4.csv", index=False)
df_temp4a
```
